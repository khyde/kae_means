```python
dataset = get_avw()
```

## Plot all maps with coastline

```python
# set dataset coordinates as lat, lon
dataset = dataset.set_coords(("lon", "lat")) 
```

```python
for i in range(len(dataset.date)):
    fig = plt.figure()
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.coastlines()
    ax.gridlines(draw_labels={"left": "y", "bottom": "x"})
    plot = dataset["avw"].isel(date = i).plot(x="lon", y="lat",vmin = 450, vmax = 575) # vmax and vmin are based on the range of "suggested_image_scaling" for max and min values of colour bar
    plot.colorbar.set_label('Apparent Visible Wavelength (nm)')
```
